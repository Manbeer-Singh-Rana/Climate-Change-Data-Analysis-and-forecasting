## **Let's do forcasting of this dataset**
- Prophet modules is used to forcast the time series data.

```python
# !pip install prophet -q
```

```python
from prophet import Prophet
```

```python
# loading forcasting model
model = Prophet()
```

```python
# data.drop('year',axis = 1 , inplace = True)
data.head()
```

```python
# for model training with prophet, we have to rename the columns
data.columns = ['ds','y']
```

```python
data.head()
```

```python
# now we can train our model with prophet 

model.fit(data)
```

```python
# now let's forcasting for next 200 data point

future = model.make_future_dataframe(200,freq = "M")
future.tail()
```

```python
future.head()
```

```python
# new prediction 
forcast = model.predict(future)[['ds', 'yhat', 'yhat_lower', 'yhat_upper']]
forcast.head()
```

```python
# data vsiualization
model.plot(forcast)
```

[Reading](https://facebook.github.io/prophet/docs/quick_start.html#python-api)

```python
from prophet.plot import plot_plotly

plot_plotly(model, forcast)
```

**making predictions a particular date**

```python
d = input("Enter the date : ")

date_in = pd.DataFrame({pd.to_datetime(d)},columns = ['ds'])
pre = model.predict(date_in)[['ds', 'yhat', 'yhat_lower', 'yhat_upper']]
pre
```

```python
plt.figure(figsize = (10,5))
plt.title(f"The Earth's tempreture at {pre.iloc[:,0]}")
sns.barplot(pre.iloc[:,1:].columns,pre.iloc[:,1:].values.reshape(-1,))
plt.show()
```
